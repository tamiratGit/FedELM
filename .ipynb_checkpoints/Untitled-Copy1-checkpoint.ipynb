```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

# Load the California Housing dataset
data = fetch_california_housing()
X, y = data.data, data.target

# Initialize the regularization parameter
alpha = 0.001

# Initialize a list to store accuracy results

def ELM_regL2(l, X, y, alpha, size):
    # Start with 50 samples and gradually increase

    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X[:size], y[:size], test_size=0.2, random_state=42)

    # Define the ELM model with L2 regularization
    #num_hidden_units = 100  # Number of hidden units
    n = X_train.shape[1]
    w = np.random.rand(n, l)
    bias = np.random.rand(1,l)

    # Calculate hidden layer outputs
    H = np.tanh(np.dot(X_train, w) + bias)

    # Calculate output weights using pseudo-inverse
    #B = np.dot(np.linalg.inv(np.eye(num_hidden_units) / reg_param + np.dot(H.T, H)), np.dot(H.T, y_train))
    B = np.dot(linalg.pinv(H.T.dot(H) + alpha * np.eye(l)), H.T.dot(y_train)) #output_weights

    # Calculate predictions on the test set
    H_test = np.tanh(np.dot(X_test, w) + bias)
    pred_test = np.dot(H_test, B)

    # Calculate RMSE (Root Mean Squared Error)
    rmse = sqrt(mean_squared_error(y_test, pred_test))
    return rmse, w ,B , bias
     
    #return mse_train, mse_test ,w ,B , b  
```

```python
# Print the accuracies for different training data sizes
l= 20# l is Number of hidden neurons
accuracies = []
for size in range(50, len(X), 50):
    rmse,_,_,_=ELM_regL2(l, X, y, alpha, size)
    accuracies.append(rmse)
```

```python
for i, accuracy in enumerate(accuracies):
    print(f"Training Data Size: {50 + i * 50}, RMSE: {accuracy:.4f}")
```
